```python
#read in all files in a directory
import os
files = 'genome files/RGI Data/allsample_loose_includenudge/text_files/'
#empty array for sample name 
sample_name = []
#empty array for sepA gene sequecen
allsepA = []
#and another 3 genes
allsdrM = []
allnorC = []
allqacj = []
for filename in os.listdir(files):
   #with open(os.path.join("files", filename), 'r') as f:
    file1 = os.path.join(files, filename)
    data = pd.read_csv(file1, sep='\t', engine='python')
    #append name to all sample names
    filenamenotxt = filename.replace('.txt','')
    #print(filenamenotxt)
    sample_name.append(filenamenotxt)
    #sample_name.append(filename)
    #add predicted sepA sequence
    sepA = np.asarray(data.loc[data['Best_Hit_ARO']=='sepA','Predicted_DNA'])
    allsepA.append(sepA)
    #and all other 3 genes
    sdrM = np.asarray(data.loc[data['Best_Hit_ARO']=='sdrM','Predicted_DNA'])
    allsdrM.append(sdrM)
    norC = np.asarray(data.loc[data['Best_Hit_ARO']=='norC','Predicted_DNA'])
    allnorC.append(norC)
    
    qacj = np.asarray(data.loc[data['Best_Hit_ARO']=='qacJ','Predicted_DNA'])
    #some species might not hit qacj
    if len(qacj)!=0:
       allqacj.append(qacj)
    else:
       allqacj.append('NA') 
    
     
    
   
```

```python
df = pd.DataFrame({'Species':sample_name, 'sepA':allsepA,'sdrM':allsdrM,'norC':allnorC,'qacJ':allqacj})
```

```python
#make super file contains all RGI's of all species

#read in all files in a directory
import os
files = 'genome files/RGI Data/allsample_loose_includenudge/text_files/'
#empty array for sample name 
sample_name = []
#empty array for sepA gene sequecen
pathogenicity = []
#strict hits
strict = 0
# create an Empty DataFrame object
df = pd.DataFrame()
for filename in os.listdir(files):
   #with open(os.path.join("files", filename), 'r') as f:
    file1 = os.path.join(files, filename)
    data = pd.read_csv(file1, sep='\t', engine='python')
    #append name to all sample names
    filenamenotxt = filename.replace('.txt','')
    sample_name.append(filenamenotxt)
    
    newdata = data[['Best_Hit_ARO','Cut_Off','Best_Identities']]
    newdata = newdata.sort_values('Best_Identities').drop_duplicates('Best_Hit_ARO', keep='last')
    #print(newdata)
    newdata = newdata.transpose()
    newdata.columns = newdata.iloc[0]
    newdata = newdata.iloc[1:]
    #print(newdata)
    #print(newdata)
    df = pd.concat([newdata,df], axis=0, ignore_index=True)
    #df.join(newdata, how='outer')

```

```python
#remove loose/strict rows
hitdf = df.iloc[1:,:]
hitdf = hitdf.iloc[::2]
hitdf = hitdf.fillna(0)
```

```python
##add species name and pathegenecity
#hitdf
hitdf['Pathogenicity'] = np.asarray(('N','Y','Y','Y','Y','N','N','N','N','Y','Y','N'))

hitdf.index = range(len(hitdf))
```

```python
x = hitdf[hitdf.columns.difference(['Pathogenicity'])]
y = hitdf['Pathogenicity']
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
pc = pca.fit_transform(x)
```

```python
pc_df = pd.DataFrame(data = pc, columns = ['PC1', 'PC2','PC3','PC4', 'PC5'])
pc_df['label'] = y
pc_df
```

```python
df = pd.DataFrame({'var':pca.explained_variance_ratio_,
             'PC':['PC1','PC2','PC3','PC4','PC5']})
sns.barplot(x='PC',y="var", 
           data=df, color="c");

for i in range(0,len(pca.explained_variance_ratio_)):
    value = pca.explained_variance_ratio_[i]
    print ("PC{}:{}".format(i+1,value))
```
